```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,50)
y1=x
y2=[i**2 for i in x ]
fig, ax= plt.subplots()
ax.set_title("y1=x,y2=x^2")
ax.grid()
ax.plot(x,y1,'--r',x,y2,'*g')
```

```python
x=np.linspace(-3,3,50)
y=x
z=np.sin(x)

rng=np.random.default_rng()
x1=rng.integers(low=-4,high=4,size=20)
y1=rng.integers(low=-4,high=4,size=20)
z1=rng.integers(low=-4,high=4,size=20)
s1 = rng.integers(low=10,high=101,size=20)

fig,ax =plt.subplots()
ax.set_title("3D Scatter Plot")
ax = plt.axes(projection='3d')
ax.grid()
ax.plot(x,y,z)

fig,ax2=plt.subplots()
ax2=plt.axes(projection='3d')
ax2.scatter(x1,y1,z1,s = s1)
```
